```python
import pandas as pd
```

```python
df=pd.read_csv('/content/sample_data/Batsman.csv')
df
```

Q-1: Against which team Virat has scored maximum runs?

```python
df.groupby(['Opposition'])['RunsScored'].sum().sort_values(ascending=False).head(1)
```

Q-2: Show the average score of Virat against each opposition team.

```python
df.groupby(['Opposition'])['RunsScored'].mean().sort_values(ascending=False)
```

Q-3: Against which team Virat has scored the maximum runs in single inning?


```python
df['RunsScored'].max()
```

```python
df[df['RunsScored']==183]
```

Q-4: In which year, Virat has scored maximum Runs.

```python
df.groupby(['Year'])['RunsScored'].sum().sort_values(ascending=False).head(1)
```

Q-5: In which year, Virat has scored min Runs.

```python
df.groupby(['Year'])['RunsScored'].sum().sort_values(ascending=True).head(1)
```

Q-6: In which year, Virat has scored maximum Runs with which opposition?

```python
df.groupby(['Year','Opposition']).RunsScored.sum().sort_values(ascending=False).head(1)
```
