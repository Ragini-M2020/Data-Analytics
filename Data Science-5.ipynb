```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Checking the list of inbuild datasets available in seaborn library
sns.get_dataset_names()
```

```python
# Loading the datasets from available inbuild datasets
df=sns.load_dataset("tips")
df
```

```python
# lmplot
# The lmplot() function plots a scatterplot and also fits a linear model with a confidence interval interval for the fitted model
sns.lmplot(data=df, x='total_bill', y='tip');
```

```python
# To check the relationship between two variables for different categories we can use hue paramter
sns.lmplot(data=df,x='total_bill',y='tip',hue='sex');
```

```python
# To plot the levels of the third variable across different plot we can use 'Col' paramter.
sns.lmplot(data=df,x='total_bill',y='tip',col='sex');
```

```python
# To  hide the confidence interval we use ci paramter and sets its value to False
sns.lmplot(data=df,x='total_bill',y='tip',col='sex',ci=False);
```

```python
# Joint Plot
# Histogram helps us to undersatnd the distribution of individual variables.
# Scatterplot help us to identify the relationship between two variables.
# Joint plot helps us plotting the histogram and scatter plot together to capture a lot more information in a concise manner
# The jointplot() function of seaborn library is used to create a jointplot
```

```python
sns.jointplot(data=df,x='total_bill',y='tip');
```

```python
sns.jointplot(data=df,x='total_bill',y='tip', kind='hex');
plt.colorbar();
```

```python
sns.jointplot(data=df,x='total_bill',y='tip', kind='kde',fill=True);
```

```python
# Violin Plot
# It allows us to visulaize the distribtuion of a numeric varaible for sevral groups
# The violinplot() function of seaboran library is used to create violin plot
# It also gies more information about the density estimation on y-axis
```

```python
sns.violinplot(data=df,x='total_bill');
```

```python
sns.violinplot(data=df,x='total_bill',y='sex');
```

```python
# To rotate the plot vertically we use 'orient' paramter
sns.violinplot(data=df,y='total_bill',x='sex',orient='v');
```

```python
# Strip Plot
# It is usefull when we have small dataset
# A strip plot is bascially a scatter plot that differntiates different categories
# The stripplot() function of seaborn library is used to create strip plot
```

```python
sns.stripplot(data=df,x='total_bill');
```

```python
sns.stripplot(data=df,x='sex',y='total_bill');
```

```python
plt.figure(figsize=(15,7))
sns.stripplot(data=df,x='sex',y='total_bill',jitter=True);
```

```python
plt.figure(figsize=(15,7))
sns.stripplot(data=df,x='sex',y='total_bill',jitter=True,hue='day');
```

```python
# Swarmplot
# It is similar to the strip plot but it avoids the overlapping of points.
# The swarmplot() function of seaborn library is use to create the swarmplot
# It plots the dots arranged in vertical columns
# It is also suitable for small datasets
```

```python
sns.swarmplot(data=df,x='sex',y='total_bill');
```

```python
sns.swarmplot(data=df,x='sex',y='total_bill',hue='day');
```

```python
sns.swarmplot(data=df,x='sex',y='total_bill',hue='day',dodge=True);
```

```python
# Cat Plot
# It provides a framework to access several types of plots and show relationship between numerical and categorical variables
# It supports 8 different plot such as stripplot, swarmplot,box, etc
# The catplot() function if seaborn libraray is use to create catplot
# Default plot is stripplot
```

```python
sns.catplot(data=df,x='sex',y='total_bill');
```

```python
sns.catplot(data=df,x='day',y='total_bill',hue='sex', kind='point');
```

```python
sns.catplot(data=df,x='day',y='total_bill',hue='sex', kind='bar');
```

```python
sns.catplot(data=df,x='day',y='total_bill',hue='sex', kind='box');
```

```python
sns.catplot(data=df,x='day',y='total_bill',hue='sex', kind='swarm');
```

```python
# Pairplot
# A pairplot shows the replationship between two numeric variables for each pair of columns in the dataset
# It create a grid of axis such that each  column has single plot
# The pairplot() function of seaborn is use to create pair plot
```

```python
sns.pairplot(data=df);
```

```python
sns.pairplot(data=df,corner=True);
```

```python
sns.pairplot(data=df,hue='sex');
```

```python
sns.pairplot(data=df,vars=['total_bill','tip'],hue='day');
```

```python
# Heatmap
# A heatmap is graphical representation of data as color-encoded matrix
# It represents the correlation for each pair of columns in the data
# The heatmap() function of seaborn is use to create heatmap
```

```python
sns.heatmap(data=df[['total_bill','tip','size']].corr());
```

```python
sns.heatmap(data=df[['total_bill','tip','size']].corr(),annot=True);
```

```python
sns.heatmap(data=df[['total_bill','tip','size']].corr(),annot=True,cbar=False);
```
