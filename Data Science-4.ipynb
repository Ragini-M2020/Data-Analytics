```python
#Histogram
#* A histogram is univariate plot which helps us understand the distribution of continuous numerical variable
#* Itr breaks the range of contionuous variables into intervals of equal length and then counts the number of observations in each interval.
#* we will use the histplot() function of seaborn to create histograms
```

```python
#To make the histogram on the price coloumn of dataset
sns.histplot(data=df,x='price');
```

```python
#Customizing the histogram
plt.title('Histogram: Price')   # To give title of graph
plt.xlim(3000,50000)            # To set the limit of x-axis
plt.ylim(0,80)                  # To set the limit of y-axis
plt.xlabel('Price of cars')     # To set the label of x-axis
plt.ylabel('Frequency')         # To set the label of y-axis
sns.histplot(data=df,x='price',color='orange',bins=20);
```

```python
sns.histplot(data=df,x='price',color='orange',binwidth=200);
# The optimal no. of bin width:
# bins=Range/binwidth
```

```python
# Kernel Density Estimation (KDE), visualizes the distribution of data over a continuous interval
sns.histplot(data=df,x='price',kde=True);
```

```python
sns.histplot(data=df,x='horsepower',kde=True);
```

```python
sns.histplot(data=df,x='horsepower',hue='body_style',kde=True);
```

```python
# The better way to plot the subplot
g=sns.FacetGrid(df,col="body_style")
g.map(sns.histplot,"horsepower");
```

```python
# Boxplot
# A boxplot or box-and-whisker plot shows the distribution of numerical data and skewness through displaying the data quartiles
# It is also called a five number summary plot -minimum value, maximum value, first quartile, meadian, thrid quartile
# The boxplot() function of seaborn is used to create boxplot
```

```python
#Bar Graph
# A bar graph is generally used to show the counts of observations in each bin(level) of categorical variable
# We can use the countplot() function of seaborn is used to create bar graph
```

```python
sns.countplot(data=df,x='body_style');
```

```python
barcontainer=sns.countplot(data=df,x='body_style')
for label in barcontainer.containers:
  barcontainer.bar_label(label)
plt.show();
```

```python
sns.countplot(data=df,x='body_style',hue='fuel_type');
```

```python
sns.countplot(data=df,x='make');
```

```python
plt.figure(figsize=(26,7))
sns.countplot(data=df,x='make');
```

```python
plt.figure(figsize=(10,7))
sns.countplot(data=df,x='make')
plt.xticks(rotation=90);
```
