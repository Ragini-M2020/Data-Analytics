The company charges are as follows:

1. 25% on the orders having final price greater than 600

2. 15% on the orders having final price greater than 350 but less than or equal to 600

3. 10% on the orders having final price greater than 100 but less than or equal to 350

4. otherwise 5% on the final price

Final price = price - discount price

Question - 5 Find the total revenue generated by E-Commerce company over all orders placed.

```python
def calculate_Revenue(final_price):
  if final_price > 600:
    return final_price*0.25
  elif final_price in range(351,601):
    return final_price*0.15
  elif final_price in range(101,351):
    return final_price*0.10
  else:
    return final_price*0.05
```

```python
#ECom_Data['Final_Price']=ECom_Data['retail_price']-ECom_Data['discounted_price']
ECom_Data['Final_Price']=ECom_Data.retail_price - ECom_Data.discounted_price
```

```python
ECom_Data['Revenue']=ECom_Data['Final_Price'].apply(calculate_Revenue)
ECom_Data.Revenue.sum()
```

Question 8: Draw a line plot for the monthly revenue of E-Commerce company for each region separately.

```python
# Converting the datatype of Order_Date column from object to datetime
ECom_Data['Order_Date']=pd.to_datetime(ECom_Data['Order_Date'], dayfirst=True)
ECom_Data.info()
```

```python
ECom_Data['Month_Year']=ECom_Data['Order_Date'].dt.month_name()+"-"+(ECom_Data['Order_Date'].dt.year).astype('string')
ECom_Data=ECom_Data.sort_values(by='Order_Date')
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=ECom_Data, x='Month_Year', y='Revenue' ,hue='Region', errorbar=('ci',False));
```
